```python
def polynomial(d):
    for i in range(d+1):
        if i==0:
            st=str(np.random.randint(1,10))
        else:
            st=st+"+"+str(np.random.randint(1,10))+"*x**"+ str(i)
    return st
```

```python
polynomial(3)
```

```python
f = 2*cos(x) + 0.3*sin(10*x)
```

```python
f
```

```python
g=integrate(f)
```

```python
g
```

```python
#把g lambdify化
h=lambdify(x, g, "numpy")
```

```python
h(3)
```

```python
t=np.linspace(0,10,1000)
y=h(t)
```

```python
plt.plot(t, y)
```
